```python
import pandas as pd
df = pd.read_csv('/content/coffee.csv')
df.head()
```

```python
# Создаём базу данных
import sqlite3
conn = sqlite3.connect('my_coffee_2.db')
```

```python
# Создаём таблицу
c = conn.cursor()
c.execute('''CREATE TABLE my_coffee_2
            (date DATE, datetime DATETIME, cash_type TEXT, card TEXT, money INTEGER, cofee_name TEXT)''')
```

```python
#сохраняем данные из csv в таблицу
for index, row in df.iterrows():
     c.execute("INSERT INTO my_coffee_2 VALUES (?, ?, ?, ?, ?, ? )", row)
```

```python
# Сохраняем изменения и закрываем соединение
conn.commit()
```

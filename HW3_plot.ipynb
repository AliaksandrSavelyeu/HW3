```python
import sqlite3
con = sqlite3.connect('/content/my_coffee.db')
cur = con.cursor()
```

```python
import pandas as pd
df = pd.read_sql('SELECT * FROM my_coffee', con)
```

```python
df.head()
```

```python
import seaborn as sns

```

```python
ax = sns.lineplot(data = df, x = 'date', y = 'money', color = 'green')
ax.set(xlabel = 'Дата', ylabel = 'Деньги')
```

```python

ax = sns.histplot(data = df, x = 'cash_type', hue = 'cash_type', palette = ['green', 'grey'])
ax.set(xlabel = 'Вид оплаты', ylabel ='Частота', title = 'Соотношение оплаты картой и наличными')
```

```python
ax = sns.histplot(data = df, y = 'cofee_name', hue = 'cofee_name', legend = False)
ax.set(xlabel = 'Количество единиц проданного кофе', ylabel = 'Наименование кофе', title = 'Количество проданных чашик кофе по каждому наименованию')
```

```python
ax = sns.histplot(data = df, x = 'money', y = 'cofee_name')
ax.set(xlabel = 'Цена за чашку кофе', ylabel = 'Наименование кофе', title = 'Цены на кофе и их встречаемость')
```

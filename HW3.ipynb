```python
import pandas as pd
df = pd.read_csv('/content/coffee.csv')
df.head()
```

```python
# Создаём базу данных
import sqlite3
conn = sqlite3.connect('my_coffee.db')
```

```python
# Создаём таблицу
c = conn.cursor()
c.execute('''CREATE TABLE my_coffee
            (date DATE, datetime DATETIME, cash_type TEXT, card TEXT, money REAL, cofee_name REAL)''')
```

```python
#сохраняем данные из csv в таблицу
for index, row in df.iterrows():
     c.execute("INSERT INTO my_coffee VALUES (?, ?, ?, ?, ?, ? )", row)
```

```python
# Сохраняем изменения и закрываем соединение
conn.commit()
```

```python

conn = sqlite3.connect('my_coffee.db')
#Запрос на все купленные Латте
query = '''SELECT *
            FROM my_coffee
            WHERE cofee_name = 'Latte' '''
#Выполнение запроса и вывод результата
for row in c.execute(query):
            print(*row)
conn.close()
```

```python
conn = sqlite3.connect('my_coffee.db')
#Запрос на всё кофе купленное за наличные в период со 2 мая 2024 по(включительно) 1 июня 2024
query = '''SELECT *
            FROM my_coffee
            WHERE cash_type = 'cash' AND date BETWEEN '2024-05-02' AND '2024-06-01' '''
#Выполнение запроса и вывод результата
for row in c.execute(query):
            print(*row)
conn.close()
```

```python
#Подсчитаем общее количество чеков
c.execute('SELECT COUNT(*) FROM my_coffee')
total_number_of_checks = c.fetchone()
print('Общее количество чеков:', total_number_of_checks)
conn.close()
```

```python
#Подсчитаем сумму полученную с продаж кофе
c.execute('SELECT SUM(money) FROM my_coffee')
amount_of_money = c.fetchone()
print('Общая сумма чеков:', amount_of_money)
conn.close()
```

```python
#Подсчитаем среднее значение чека
c.execute('SELECT AVG(money) FROM my_coffee')
average_bill = c.fetchone()
print('Средний чек:', average_bill)
conn.close()
```

```python
#Максимальный чек
c.execute('SELECT MAX(money) FROM my_coffee')
max_bill = c.fetchone()
print('Максимальный чек:', max_bill)
conn.close()
```

```python
#Минимальный чек
c.execute('SELECT MIN(money) FROM my_coffee')
min_bill = c.fetchone()
print('Минимальный чек:', min_bill)
conn.close()
```
